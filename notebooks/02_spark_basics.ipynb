```python
def crazy_stuff(base_num):
    num = int(base_num[1:])
    if num % 7 == 0:
        return f's/{num:03x}'
    else:
        return f'e/{num:03x}'
```

```python
crazy_stuff('B02884')
```

```python
crazy_stuff_udf = F.udf(crazy_stuff, returnType=types.StringType())
```

```python
df \
    .withColumn('crazy',crazy_stuff_udf(df.dispatching_base_num)) \
    .select('crazy','dispatching_base_num') \
    .show()
```
